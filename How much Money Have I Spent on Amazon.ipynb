# Cleaning The Data

```python
Amazon_data = Amazon_data.fillna(0)
```

```python
Amazon_data["Total Charged"] = Amazon_data["Total Charged"].str.replace('$','').astype(float) 
```

## Analyzing my Amazon Spending over time 

```python
Amazon_data["Order Date"] = pd.to_datetime(Amazon_data["Order Date"])
```

```python
Amazon_data
```

```python
Amazon_data.plot.bar(x='Order Date',y='Total Charged', rot=90, figsize=(20,10))
plt.title('Plot of Total charge by order for various order dates')
plt.xlabel('Order date')
plt.ylabel('Total Charged')
plt.tight_layout()
```

```python
daily_orders = Amazon_data.groupby('Order Date').sum()['Total Charged']
daily_orders
```

```python
daily_orders.plot.bar(figsize=(20,10), color='#61D199')
plt.title('Plot of Total charge by order by day')
plt.xlabel('Order date')
plt.ylabel('Total Charged')
plt.tight_layout()
```
